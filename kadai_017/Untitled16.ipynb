```python
class Human:
    def __init__(self, name , age,):
         self.name = name
         self.age = age
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")


humans = []
humans.append(Human("1",20))
humans.append(Human("2",18))
humans.append(Human("3",23))

for human  in humans:
    human.check_adult()
```
